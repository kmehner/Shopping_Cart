Build a shopping cart program with the following capabilities:

1) Takes in an input

2) Stores user input into a dictionary or list

3) The User can add or delete items

4) The User can see current shopping list

5) The program Loops until user 'quits'

6) Upon quiting the program, prints out a receipt of the items with total and quantity


```python
from IPython.display import clear_output

def shoppinglist():
    mydict = {}
    count = 0
    
    takeinput = True
    
    print("Today we will create your shopping cart.")
    
    while takeinput:
        price_num = {}
        
        # Displaying items currently in the cart
        for key, value in mydict.items(): 
                print(f"{value['num']} {key}s x {value['price']} ")
                
        # Asking the user what they would like to do        
        repeat = input("What would you like to do? (Add/Remove/Edit/Clear/Quit): ").lower()
        while repeat not in ['add', 'remove', 'edit', 'clear', 'quit']:
            repeat = input('That is not a valid response. Please type Add, Remove, Edit, Clear or Quit)').lower()
        
        # Letting the user add to their shopping list 
        if repeat == 'add':
            item = input("What would you like to add?: ").title()
            
            # Checking that the item is not already on their shopping list 
            while item in mydict.keys():
                item = input("Sorry, this item is already on the shopping list. What other item would you like to add?: ")
            
            # Getting the price and saving to nested_dict (must be able to be converted to float for total)
            price = input(f"What is the price of {item}: ").title()
            try:
                float(price)
            except ValueError:
                price = input(f"Please enter a numeric value. What is the price of {item}: ")
            price_num['price'] = price
            
            # Getting the quantity and saving to nested_dict (must be able to be converted to float for total)
            num = input(f"How many {item}s would you like to add: ").title()
            try:
                float(num)
            except ValueError:
                    num = input(f"Please enter a numeric value. What is the quantity of {item}: ")
            price_num['num'] = num
            
            # Saving to dictionary
            mydict[item] = price_num
            
            clear_output()
            print("Items in shopping list: ")
        
        # remove
        elif repeat == 'remove':
            removethis = input("What would you like to remove?: ").title()
            del mydict[removethis]
            
        # edit 
        elif repeat == 'edit':
            clear_output()
            
            # Providing options to edit
            print("You currently have: ")
            for key, value in mydict.items(): 
                print(f"{value['num']} {key}s x {value['price']} ")
            
            # prompt for item to edit - must be existing item
            editthis = input("What would you like to edit?: ").title()
            while editthis not in mydict.keys():
                editthis = input('That is not a valid response. What would you like to edit from your current list?: ').title()
                if editthis == 'quit':
                    return None
            
            # New price and quantity 
                # new variables to avoiding all items with same price/quantity
                # checking to see if input can be converted to a float 
            
            newprice = input("What is the updated price?: ")
            try:
                float(newprice)
            except ValueError:
                newprice = input(f"Please enter a numeric value. What is the updated price of {item}: ")
            price_num['price'] = newprice
            
            newnum = input("What is the updated quantity? ")
            price_num['num'] = newnum
            try:
                int(newnum)
            except ValueError:
                newnum = input(f"Please enter a numeric value. What is the quantity price of {item}")
            
            # Saving to dictionary
            mydict[editthis] = price_num
            
            clear_output()
            print("Items in shopping list: ")
        
        # Clear 
        elif repeat == 'clear':
            clear_output()
            
        # Quit
        elif repeat == 'quit':
            takeinput = False
        
 
    clear_output()
    
    # Getting count for total (includes updated price if item was edited)
    for key, value in mydict.items(): 
            count += int(value['num']) * int(value['price'])
    
    # rounding total to two decimal places
    answer = str(round(count, 2))
            
    print("\n - - - - - Final Receipt - - - - - \n")
    
    # Listing quantity, item, and price
    for key, value in mydict.items(): 
            print(f"{value['num']} {key}s x {value['price']} \n")
    
    # Total price and thank you 
    print(f"The total is {answer}\n")
    print(" - - - - - - - - End - - - - - - - \n  Thank you for shopping with us!")


shoppinglist()
```
